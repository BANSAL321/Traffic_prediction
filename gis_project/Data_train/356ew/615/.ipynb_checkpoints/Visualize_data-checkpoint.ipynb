```python
import numpy as np
import matplotlib.pyplot as plt

for fname in ('frequency1.txt', 'frequency2.txt' ...):
    data=np.loadtxt(fname)
    X=data[:,1]
    Y=data[:,2]
    plt.plot(X,Y,':ro')
plt.ylim((0,55000))
plt.show() 
plt.save('figure.png')
```

```python
import numpy as np
import matplotlib.pyplot as plt

filelist=[]

for i in range(1,31):
#     filelist.append("frequency%s.dat" %i)
    filelist.append("356_EW_%sjuly_150220615" %i)
#     356_EW_10july_150220615

for fname in filelist:
    data=np.loadtxt(fname)
    X=data[:,0]
    Y=data[:,1]
    plt.plot(X,Y,':ro')

plt.show()
```
